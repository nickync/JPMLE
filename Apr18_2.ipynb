```python
import pandas as pd
```

```python
df = pd.read_csv('https://raw.githubusercontent.com/fenago/MLEssentials2/main/datasets/adult-census.csv')
```

```python
df.head()
```

```python
df.sample(5)
```

```python
target_name = "class"
target = df[target_name]
target
```

```python
data = df.drop(columns=[target_name])
data.head()
```

```python
data = df.drop(columns=['workclass', 'education', 'education-num','marital-status','occupation','relationship','race','sex','native-country','class'])
data.head()
```

```python
print(f"The dataset contains {data.shape[0]} samples and "
      f"{data.shape[1]} features")
```

```python
from sklearn.neighbors import KNeighborsClassifier
model = KNeighborsClassifier()
model.fit(data, target)
```

```python
target_predicted = model.predict(data)
```

```python
target_predicted[:5]
```

```python
target_predicted[:5] == target[:5]
```

```python
print(f"Number of correct prediction: "
      f"{(target[:5] == target_predicted[:5]).sum()} / 5")
```
